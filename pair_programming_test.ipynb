```python
from cmath import cos, sin
r = float(12) #radius to be used for conversion
teeta = float(30) #angle in degrees to be used for conversion
teeta_rad = teeta*(3.14159/180) 
x = r*cos(teeta_rad) 
y = r*sin(teeta_rad) 
print(f"The polar coordinates (r={r}, θ={teeta}°) are (x={x}, y={y}) in cartesian coordinates.")
#the polar coordinates should print as (r=12.0, θ=30.0°) and the cartesian coordinates should print as (x=10.39, y=6)
```
